Написать программу, которая будет считывать данные из CSV файла, содержащего
информацию о продажах товаров в магазине. Данные в файле содержатся в следующем
формате:
| Номер заказа | Дата заказа | Название товара | Категория товара | Количество продаж | Цена
за единицу | Общая стоимость |
Необходимо:
1. Рассчитать общую выручку магазина.
2. Найти товар, который был продан наибольшее количество раз.
3. Найти товар, который принес наибольшую выручку.
4. Составить отчет, содержащий информацию об общей выручке магазина, количестве
проданных единиц каждого товар и доле каждого товара в общей выручке.

```python
import csv
import pandas as pd
```

```python
df = pd.read_csv("D:/table2.csv", delimiter=';', index_col='Номер заказа')
df
```

```python
profit = 0
sales = 0
mostFrequentProduct = {}
highestProfitProduct = {}
productCategories = {}

with open("D:/table2.csv", mode="r", encoding="utf-8") as f:
    #Номер заказа; Дата заказа; Название товара; Категория товара; Количество продаж; Цена за единицу; Общая стоимость
    reader = csv.reader(f, delimiter=";")
    for row in reader:
        try:
            profit += int(row[-1])
            sales += int(row[-3])

            if row[3] not in productCategories:
                productCategories[row[3]] = [int(row[-3]), int(row[-1])]
            else:
                productCategories[row[3]][0] += int(row[-3])
                productCategories[row[3]][1] += int(row[-1])

            if row[2] not in mostFrequentProduct:
                mostFrequentProduct[row[2]] = int(row[-3])

            if row[2] not in highestProfitProduct:
                highestProfitProduct[row[2]] = int(row[-1])
        except:
            pass
```

```python
#Общая выручка
print(f"Общая выручка магазина = {profit} руб, всего товаров продано = {sales} шт")
#Самый частый
mostFrequent = [k for k, v in mostFrequentProduct.items() if v == max(mostFrequentProduct.values())]
items = ", ".join(mostFrequent)
print(f"Наибольшие продажи у товаров: {items}, {max(mostFrequentProduct.values())} штук")
#Самый прибыльный
highestProfit = [k for k, v in highestProfitProduct.items() if v == max(highestProfitProduct.values())]
items = ", ".join(highestProfit)
print(f"Наибольшая выручка у товаров: {items}, {max(highestProfitProduct.values())} руб")
#Сводная таблица по категориям (количество, обшая сумма, доля в выручке)
percent = []
for prod in productCategories:
    percent.append(productCategories[prod][1]/profit*100)
df = pd.DataFrame(productCategories, index = ["Количество проданных единиц, шт", "Общая сумма товаров, руб"])
df.loc["Доля товара в общей выручке, %"] = percent
df.transpose().sort_values("Общая сумма товаров, руб", ascending=False).round(2)
```
