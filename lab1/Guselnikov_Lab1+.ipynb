```python
import time
import copy
import math
from numpy.random import seed
from numpy.random import randint
import matplotlib.pyplot as plt
```

```python
#генерация случайных списков разной длины
matrix =  [] #коллекция списков
for i in range(9):
    a = randint(0, 1000 * (i+1), 1000 * (i+1))
    matrix.append(a)
```

### Задание
Реализовать методы сортировки строк числовой матрицы в соответствии с
заданием. Построить графики зависимости времени выполнения алгоритмов сортировки от n

```python
# Сортировка выбором.
# Берётся срез массива, в котором минимальный элемент переносят в самый левый угол,
# после чего срез уменьшается и цикл повторяется.
#TODO
def rjaka_sort(a):
    s = []
    while len(a)!=0:
        m = min(a)
        s.append(m)
        a.remove(m)
    return s

elements_rjaka = list()
times_rjaka = list()

for k in range(len(matrix)):
    a = copy.copy(matrix[k])
    start = time.perf_counter()
    rjaka_sort(list(a))
    end = time.perf_counter()
    print(len(a), "Элементов, отсортированных по rjaka сортировке за", round(end-start, 2), "cекунд")
    elements_rjaka.append(len(a))
    times_rjaka.append(end-start)

plt.xlabel('Длина списка')
plt.ylabel('Временная сложность')
plt.plot(elements_rjaka, times_rjaka, label ='rjaka sort')
plt.grid()
plt.legend()
plt.show() 
```

```python
# Сортировка выбором.
# Берётся срез массива, в котором минимальный элемент переносят в самый левый угол,
# после чего срез уменьшается и цикл повторяется.
#TODO
def Selection_sort(a):
    for i in range(len(a) - 1):
        min = i
        for j in range(i + 1, len(a)):
            if a[j] < a[min]:
                min = j
        a[i], a[min] = a[min], a[i]
    return a

elements_select = list()
times_select = list()

for k in range(len(matrix)):
    a = copy.copy(matrix[k])
    start = time.perf_counter()
    Selection_sort(a)
    end = time.perf_counter()
    print(len(a), "Элементов, отсортированных по выборочной сортировке за", round(end-start, 2), "cекунд")
    elements_select.append(len(a))
    times_select.append(end-start)

plt.xlabel('Длина списка')
plt.ylabel('Временная сложность')
plt.plot(elements_select, times_select, label ='Selection sort')
plt.grid()
plt.legend()
plt.show() 
```

```python
# Сортировка вставкой.
#TODO
def Insertion_sort(a):
    for i in range(1, len(a)):
        for j in range(i, 0, -1):
            if a[j] < a[j-1]:
                a[j], a[j-1] = a[j-1], a[j]
            else:
                break
    return a

elements_insert = list()
times_insert = list()

for k in range(len(matrix)):
    a = copy.copy(matrix[k])
    start = time.perf_counter()
    Insertion_sort(a)
    end = time.perf_counter()
    print(len(a), "Элементов, отсортированных по сортировке вставкой за", round(end-start, 2), "cекунд")
    elements_insert.append(len(a))
    times_insert.append(end-start)

plt.xlabel('Длина списка')
plt.ylabel('Временная сложность')
plt.plot(elements_insert, times_insert, label ='Insertion sort')
plt.grid()
plt.legend()
plt.show()    
```

```python
#Сортировка пузырьком.
#TODO
def Bubble_sort(a):
    for i in range(len(a)-1):
        for j in range(len(a)-1-i):
            if a[j] > a[j+1]:
                a[j], a[j+1] = a[j+1], a[j]
    return a

elements_bubble = list()
times_bubble = list()

for k in range(len(matrix)):
    a = copy.copy(matrix[k])
    start = time.perf_counter()
    Bubble_sort(a)
    end = time.perf_counter()
    print(len(a), "Элементов, отсортированных по пузырьковой сортировке за ", round(end-start, 2), "секунд")
    elements_bubble.append(len(a))
    times_bubble.append(end-start)

plt.xlabel('Длина списка')
plt.ylabel('Временная сложность')
plt.plot(elements_bubble, times_bubble, label ='Bubble sort')
plt.grid()
plt.legend()
plt.show()      
```

```python
# Сортировка Шелла.
#TODO
def Shell_sort(a):
    gap = len(a) // 2
    while gap > 0:
        for i in range(gap, len(a)):
            cur_val = a[i]
            pos = i
            while pos >= gap and a[pos - gap] > cur_val:
                a[pos] = a[pos - gap]
                pos -= gap
                a[pos] = cur_val
        gap //= 2
    return a

elements_shell = list()
times_shell = list()

for k in range(len(matrix)):
    a = copy.copy(matrix[k])
    start = time.perf_counter()
    Shell_sort(a)
    end = time.perf_counter()
    print(len(a), "Элементов, отсортированных по сортировке Шелла за", round(end-start, 4), "cекунд")
    elements_shell.append(len(a))
    times_shell.append(end-start)

plt.xlabel('Длина списка')
plt.ylabel('Временная сложность')
plt.plot(elements_shell, times_shell, label ='Shell sort')
plt.grid()
plt.legend()
plt.show()    
```

```python
# Быстрая сортировка.
def quick_sort(a):
    if len(a) <= 1:
        return a

    greater = []
    equal = []
    lesser = []
    for i in range(len(a)):
        if a[i] == a[0]:
            equal.append(a[i])
        elif a[i] > a[0]:
            greater.append(a[i])
        else:
            lesser.append(a[i])
    return quick_sort(lesser) + equal + quick_sort(greater)

elements_quick = list()
times_quick = list()

for k in range(len(matrix)):
    a = copy.copy(matrix[k])
    start = time.perf_counter()
    quick_sort(a)
    end = time.perf_counter()
    print(len(a), "Элементов, отсортированных по быстрой сортировке за", round(end-start, 4), "cекунд")
    elements_quick.append(len(a))
    times_quick.append(end-start)

plt.xlabel('Длина списка')
plt.ylabel('Временная сложность')
plt.plot(elements_quick, times_quick, label ='Quick sort')
plt.grid()
plt.legend()
plt.show()      
```

```python
# Турнирная сортировка(Пирамидальная).
#TODO
def heapify(a, n, i):
    max = i
    left = 2*i+1
    right = 2*i+2
    if left < n and a[i] < a[left]:
        max = left
    if right < n and a[max] < a[right]:
        max = right

    if max != i:
        a[i], a[max] = a[max], a[i]
        heapify(a, n, max)

def heap_sort(a):
    for i in range(len(a)//2, -1, -1):
        heapify(a, len(a), i)
    for i in range(len(a)-1, 0, -1):
        a[i], a[0] = a[0], a[i]
        heapify(a, i, 0)
    return a

elements_heap = list()
times_heap = list()

for k in range(len(matrix)):
    a = copy.copy(matrix[k])
    start = time.perf_counter()
    heap_sort(a)
    end = time.perf_counter()
    print(len(a), "Элементов, отсортированных по турнирной сортировке за", round(end-start, 4), "cекунд")
    elements_heap.append(len(a))
    times_heap.append(end-start)

plt.xlabel('Длина списка')
plt.ylabel('Временная сложность')
plt.plot(elements_heap, times_heap, label ='Heap sort')
plt.grid()
plt.legend()
plt.show()
```
